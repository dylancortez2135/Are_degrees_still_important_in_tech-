```python
# IMPORTING LIBRARIES

import pandas as pd
import numpy as np
```

```python
# CREATING DATAFRAMES
pd.set_option('display.max.columns', 30)
pd.set_option('display.max.rows', 1300)
employee_df = pd.read_csv(r"C:\Users\Admin\OneDrive\Pictures\dylan profile\dylan gr11\Dylan\Data analyst Dylan\PROJECT DEGREES\Employee.csv")
performance_df = pd.read_csv(r"C:\Users\Admin\OneDrive\Pictures\dylan profile\dylan gr11\Dylan\Data analyst Dylan\PROJECT DEGREES\PerformanceRating.csv")
```

```python
# GETTING THE AVERAGE SELF & MANAGER RATINGS.

ave_self_rating = performance_df.groupby('EmployeeID')['SelfRating'].mean()
ave_manager_rating = performance_df.groupby('EmployeeID')['ManagerRating'].mean()
```

```python
# REMOVING THE EMPLOYEEID DUPLICATES TO ASSIGN THE AVERAGE RATINGS TO.

performance_df = performance_df.drop_duplicates(subset='EmployeeID')
```

```python
# ADDING THE AVERAGE RATINGS FROM THE performance_df TO THE employee_df.

employee_performance_df = pd.merge(pd.merge(employee_df, ave_self_rating, on = 'EmployeeID', how = 'inner'), 
                                   ave_manager_rating, on = 'EmployeeID', how = 'inner')
```

```python
# ASSIGNING CATEGORIES TO THE RATINGS
rating_categories = ['Excellent', 'Above Average', 'Average', 'Below Average', 'Poor']


self_rating_conditions = [employee_performance_df['SelfRating'] >= 4.5,
                    employee_performance_df['SelfRating'] >= 3.5,
                    employee_performance_df['SelfRating'] >= 2.5,
                    employee_performance_df['SelfRating'] >= 1.5,
                    employee_performance_df['SelfRating'] < 1.5,]

manager_rating_conditions = [employee_performance_df['ManagerRating'] >= 4.5,
                    employee_performance_df['ManagerRating'] >= 3.5,
                    employee_performance_df['ManagerRating'] >= 2.5,
                    employee_performance_df['ManagerRating'] >= 1.5,
                    employee_performance_df['ManagerRating'] < 1.5,]

employee_performance_df['self_rating_category'] = np.select(self_rating_conditions, rating_categories)
employee_performance_df['manager_rating_category'] = np.select(manager_rating_conditions, rating_categories)
```

```python
# ASSIGNING EDUCATION VALUES

education_level_conditions = [employee_performance_df['Education'] == 1,
                              employee_performance_df['Education'] == 2, 
                              employee_performance_df['Education'] == 3,
                              employee_performance_df['Education'] == 4,
                              employee_performance_df['Education'] == 5]

education_level_categories = ['No Formal Qualifications', 'High School', "Bachelor's Degree", "Master's Degree", 'PhD']

education_level = np.select(education_level_conditions, education_level_categories)
```

```python
employee_performance_df['Education'] = np.select(education_level_conditions, education_level_categories)
```

```python
# CREATING A EmployeeName Column

employee_performance_df['EmployeeName'] = employee_performance_df['FirstName'] + ' ' + employee_performance_df['LastName']
```

```python
# CREATING A FINAL DATAFRAME CONTAINING ONLY THE RELEVANT COLUMNS

employee_performance_df_final = pd.merge(employee_performance_df[['EmployeeID', 'EmployeeName', 'Education', 'EducationField', 'Department', 'JobRole', 
                         'SelfRating', 'self_rating_category', 'ManagerRating', 'manager_rating_category', 'Salary', 
                         'MaritalStatus', 'Ethnicity', 'Gender', 'Age', 'YearsAtCompany']], performance_df[['EmployeeID', 'TrainingOpportunitiesTaken']], 
         on = 'EmployeeID', how = 'left')
```

```python
# CREATING AN EDUCATION LEVEL SORTER FOR CREATING CLEAN VISUALIZATIONS

education_level_sorter_conditions = [employee_performance_df_final['Education'] == 'PhD',
                                     employee_performance_df_final['Education'] == "Master's Degree",
                                     employee_performance_df_final['Education'] == "Bachelor's Degree",
                                     employee_performance_df_final['Education'] == 'High School',
                                     employee_performance_df_final['Education'] == 'No Formal Qualifications']

education_level_sorter_categories = [1, 2, 3, 4, 5]



employee_performance_df_final['education_sorter'] = np.select(education_level_sorter_conditions, education_level_sorter_categories)
```

```python
# FINAL CLEANING

employee_performance_df_final.loc[
    employee_performance_df_final['EducationField'].str.contains('Marketing', na=False),
    'EducationField'
] = 'Marketing'
```
